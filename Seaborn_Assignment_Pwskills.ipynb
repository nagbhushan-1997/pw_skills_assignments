Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Scatter Plot: A scatter plot helps us understand the relationship between two continuous variables. It displays individual data points as dots, allowing us to identify patterns, correlations, or outliers.
Example use case: Analyzing the correlation between a student’s study hours and their exam scores.
Line Plot: Line plots are useful for visualizing trends over time or across ordered categories. They connect data points with lines, making it easy to observe changes.
Example use case: Showing the stock price movement over several months.
Strip Plot: A strip plot displays individual data points for categorical variables along a single axis. It’s helpful for comparing distributions within different categories.
Example use case: Comparing the spending amounts on different days of the week.
Swarm Plot: Similar to a strip plot, a swarm plot avoids overlapping data points by jittering them. It provides a clearer view of the distribution.
Example use case: Visualizing the sepal length of different iris species.
Box Plot (Box-and-Whisker Plot): Box plots summarize the distribution of a continuous variable within different categories. They show median, quartiles, and potential outliers.
Example use case: Comparing the distribution of house prices across different neighborhoods.
Histogram: Histograms display the frequency distribution of a continuous variable. They group data into bins and show how data is distributed.
Example use case: Analyzing the distribution of ages in a population.
Kernel Density Plot: Kernel density plots estimate the probability density function of a continuous variable. They provide a smooth representation of the data distribution.
Example use case: Understanding the distribution of body mass index (BMI) in a sample.



Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.
Note: timepoint, signal, event, and region are columns in the fmri dataset.

```python
import seaborn as sns
```

```python
fmri = sns.load_dataset('fmri')
```

```python
fmri.head()
```

```python
sns.lineplot(x="timepoint", y="signal", hue="region", style="event" , data = fmri)
```
